```python
# Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
# Importing Dataset
books=pd.read_csv('book.csv')
books.head()
```

```python
# Data is already in transaction format
```

### Apriori Algorithm

#### 1) Association rules with 10% support and 70% confidence

```python
frequent_itemsets=apriori(books,min_support=0.1,use_colnames=True)
frequent_itemsets
```

```python
rules=association_rules(frequent_itemsets,metric='lift',min_threshold=0.7)
rules
```

```python
# Leverage value of 0 indicates independence. Range will be [-1 to 1]
# High conviction value indicates that the consequent is highly depending on the antecedant 
```

```python
rules.sort_values('lift',ascending=False)
```

```python
# Lift Ratio > 1 is a good influential rule in selecting the associated transactions
rules[rules.lift>1]
```

```python
# Visualization for obtained rule
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('support')
plt.xlabel('confidence')
plt.show()
```
